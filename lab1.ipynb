```python
dota = pd.read_csv("/content/sample_data/dota2HeroPick.csv")
dota = pd.get_dummies(dota, columns=['Primary Attribute', 'Attack Type', 'Niche Hero?'], prefix=['PA', 'AT', 'Niche'])

all_roles = set()
for roles in dota['Roles'].dropna():
    for role in roles.split(', '):
        all_roles.add(role.strip())

for role in all_roles:
    dota[f'Role_{role}'] = dota['Roles'].apply(lambda x: 1 if role in str(x) else 0)

dota = dota.drop(columns=['Roles'])

#print(dota.head())
dota
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
sns.set_context("notebook", font_scale=1.2)

dota = pd.read_csv("/content/sample_data/dota2HeroPick.csv", index_col=0)

df_viz = dota.copy()

all_roles = set()
for roles in df_viz['Roles'].dropna():
    for role in roles.split(', '):
        all_roles.add(role.strip())

for role in all_roles:
    df_viz[f'Role_{role}'] = df_viz['Roles'].apply(lambda x: 1 if role in str(x) else 0)

fig = plt.figure(figsize=(20, 30))

ax1 = plt.subplot(6, 3, 1)
primary_counts = df_viz['Primary Attribute'].value_counts()
colors = ['#ff6b6b', '#4ecdc4', '#45b7d1', '#96ceb4']
wedges, texts, autotexts = ax1.pie(primary_counts.values, labels=primary_counts.index,
                                    autopct='%1.1f%%', colors=colors, startangle=90,
                                    explode=[0.05]*4, shadow=True)
ax1.set_title('Распределение героев по первичным атрибутам', fontsize=14, fontweight='bold')

ax2 = plt.subplot(6, 3, 2)
sns.barplot(x=primary_counts.index, y=primary_counts.values, palette='viridis', ax=ax2)
ax2.set_title('Количество героев по атрибутам', fontsize=14, fontweight='bold')
ax2.set_xlabel('Первичный атрибут')
ax2.set_ylabel('Количество героев')

for i, v in enumerate(primary_counts.values):
    ax2.text(i, v + 2, str(v), ha='center', fontweight='bold')

ax3 = plt.subplot(6, 3, 4)
attack_counts = df_viz['Attack Type'].value_counts()
colors = ['#ff9999', '#66b3ff']
wedges, texts, autotexts = ax3.pie(attack_counts.values, labels=attack_counts.index,
                                    autopct='%1.1f%%', colors=colors, startangle=90,
                                    explode=[0.03, 0.03], shadow=True)
ax3.set_title('Тип атаки: Melee vs Ranged', fontsize=14, fontweight='bold')

ax4 = plt.subplot(6, 3, 5)
sns.barplot(x=attack_counts.index, y=attack_counts.values, palette='coolwarm', ax=ax4)
ax4.set_title('Распределение по типу атаки', fontsize=14, fontweight='bold')
ax4.set_xlabel('Тип атаки')
ax4.set_ylabel('Количество героев')
for i, v in enumerate(attack_counts.values):
    ax4.text(i, v + 2, str(v), ha='center', fontweight='bold')

ax5 = plt.subplot(6, 3, 7)
niche_counts = df_viz['Niche Hero?'].value_counts()
colors = ['#98d8c8', '#f4a582']
wedges, texts, autotexts = ax5.pie(niche_counts.values, labels=['Не нишевый', 'Нишевый'],
                                    autopct='%1.1f%%', colors=colors, startangle=90,
                                    explode=[0.02, 0.05], shadow=True)
ax5.set_title('Нишевые герои', fontsize=14, fontweight='bold')

ax6 = plt.subplot(6, 3, 8)
sns.barplot(x=['Не нишевый', 'Нишевый'], y=niche_counts.values, palette='Set2', ax=ax6)
ax6.set_title('Распределение нишевых героев', fontsize=14, fontweight='bold')
ax6.set_xlabel('Тип героя')
ax6.set_ylabel('Количество')
for i, v in enumerate(niche_counts.values):
    ax6.text(i, v + 2, str(v), ha='center', fontweight='bold')

# Топ-10 героев по победам
ax7 = plt.subplot(6, 3, (10, 11))
top_wins = df_viz.nlargest(10, 'Total Pro wins')[['Name', 'Total Pro wins']]
sns.barplot(data=top_wins, y='Name', x='Total Pro wins', palette='rocket_r', ax=ax7)
ax7.set_title('Топ-10 героев по победам на про-сцене', fontsize=14, fontweight='bold')
ax7.set_xlabel('Количество побед')
ax7.set_ylabel('Герой')

# Топ-10 пик
ax8 = plt.subplot(6, 3, (13, 14))
top_picked = df_viz.nlargest(10, 'Times Picked')[['Name', 'Times Picked']]
sns.barplot(data=top_picked, y='Name', x='Times Picked', palette='mako_r', ax=ax8)
ax8.set_title('Топ-10 самых пикаемых героев', fontsize=14, fontweight='bold')
ax8.set_xlabel('Количество пиков')
ax8.set_ylabel('Герой')

# Топ-10 самых героев, в бан которых отправить не жалко
ax9 = plt.subplot(6, 3, (16, 17))
top_banned = df_viz.nlargest(10, 'Times Banned')[['Name', 'Times Banned']]
sns.barplot(data=top_banned, y='Name', x='Times Banned', palette='magma_r', ax=ax9)
ax9.set_title('Топ-10 самых банаемых героев', fontsize=14, fontweight='bold')
ax9.set_xlabel('Количество банов')
ax9.set_ylabel('Герой')

plt.tight_layout()
plt.show()

# Корелляция
fig2, (ax10, ax11) = plt.subplots(1, 2, figsize=(18, 7))

# Выбрали столбики
numeric_cols = ['Total Pro wins', 'Times Picked', 'Times Banned', 'Win Rate']
corr_matrix = df_viz[numeric_cols].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8}, ax=ax10)
ax10.set_title('Корреляция между метриками героев', fontsize=14, fontweight='bold')

# Win rate
ax11.hist(df_viz['Win Rate'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
ax11.axvline(df_viz['Win Rate'].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Среднее: {df_viz["Win Rate"].mean():.1f}%')
ax11.axvline(df_viz['Win Rate'].median(), color='green', linestyle='dashed', linewidth=2, label=f'Медиана: {df_viz["Win Rate"].median():.1f}%')
ax11.set_title('Распределение процента побед героев', fontsize=14, fontweight='bold')
ax11.set_xlabel('Win Rate (%)')
ax11.set_ylabel('Количество героев')
ax11.legend()
ax11.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

fig3, axes = plt.subplots(2, 2, figsize=(18, 14))

ax12 = axes[0, 0]
sns.boxplot(data=df_viz, x='Primary Attribute', y='Win Rate', palette='Set3', ax=ax12)
ax12.set_title('Распределение Win Rate по первичным атрибутам', fontsize=14, fontweight='bold')
ax12.set_xlabel('Первичный атрибут')
ax12.set_ylabel('Win Rate (%)')
ax12.grid(True, alpha=0.3)

ax13 = axes[0, 1]
sns.violinplot(data=df_viz, x='Attack Type', y='Win Rate', palette='pastel', ax=ax13)
ax13.set_title('Распределение Win Rate по типу атаки', fontsize=14, fontweight='bold')
ax13.set_xlabel('Тип атаки')
ax13.set_ylabel('Win Rate (%)')
ax13.grid(True, alpha=0.3)

ax14 = axes[1, 0]
sns.regplot(data=df_viz, x='Times Picked', y='Total Pro wins',
            scatter_kws={'alpha':0.6, 's':50}, line_kws={'color':'red'}, ax=ax14)
ax14.set_title('Зависимость побед от количества пиков', fontsize=14, fontweight='bold')
ax14.set_xlabel('Количество пиков')
ax14.set_ylabel('Количество побед')
ax14.grid(True, alpha=0.3)

# Топ-10 героев по винрейту
ax15 = axes[1, 1]
top_winrate = df_viz.nlargest(10, 'Win Rate')[['Name', 'Win Rate']]
sns.barplot(data=top_winrate, y='Name', x='Win Rate', palette='Spectral_r', ax=ax15)
ax15.set_title('Топ-10 героев по Win Rate', fontsize=14, fontweight='bold')
ax15.set_xlabel('Win Rate (%)')
ax15.set_ylabel('Герой')
for i, (_, row) in enumerate(top_winrate.iterrows()):
    ax15.text(row['Win Rate'] + 0.5, i, f'{row["Win Rate"]:.1f}%', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

fig4, axes = plt.subplots(2, 2, figsize=(18, 14))

role_columns = [col for col in df_viz.columns if col.startswith('Role_')]
role_stats = []

for role_col in role_columns:
    role_name = role_col.replace('Role_', '')
    heroes_with_role = df_viz[df_viz[role_col] == 1]
    role_stats.append({
        'Role': role_name,
        'Count': len(heroes_with_role),
        'Avg_WinRate': heroes_with_role['Win Rate'].mean(),
        'Total_Wins': heroes_with_role['Total Pro wins'].sum(),
        'Total_Picks': heroes_with_role['Times Picked'].sum()
    })

role_df = pd.DataFrame(role_stats).sort_values('Count', ascending=False)

ax16 = axes[0, 0]
sns.barplot(data=role_df.head(10), y='Role', x='Count', palette='viridis', ax=ax16)
ax16.set_title('Топ-10 ролей по количеству героев', fontsize=14, fontweight='bold')
ax16.set_xlabel('Количество героев')
ax16.set_ylabel('Роль')

ax17 = axes[0, 1]
sns.barplot(data=role_df.head(10), y='Role', x='Avg_WinRate', palette='plasma', ax=ax17)
ax17.set_title('Средний Win Rate по ролям (топ-10)', fontsize=14, fontweight='bold')
ax17.set_xlabel('Средний Win Rate (%)')
ax17.set_ylabel('Роль')
for i, (_, row) in enumerate(role_df.head(10).iterrows()):
    ax17.text(row['Avg_WinRate'] + 0.2, i, f'{row["Avg_WinRate"]:.1f}%', va='center')

ax18 = axes[1, 0]
top_5_roles = role_df.head(5)
other_roles_count = role_df['Count'][5:].sum()
sizes = list(top_5_roles['Count']) + [other_roles_count]
labels = list(top_5_roles['Role']) + ['Другие']
colors = plt.cm.Set3(np.linspace(0, 1, 6))
wedges, texts, autotexts = ax18.pie(sizes, labels=labels, autopct='%1.1f%%',
                                     colors=colors, startangle=90, shadow=True)
ax18.set_title('Распределение героев по основным ролям', fontsize=14, fontweight='bold')

ax19 = axes[1, 1]
role_attr_matrix = []
for role in top_5_roles['Role']:
    row = []
    for attr in ['str', 'agi', 'int', 'all']:
        count = len(df_viz[(df_viz[f'Role_{role}'] == 1) & (df_viz['Primary Attribute'] == attr)])
        row.append(count)
    role_attr_matrix.append(row)

sns.heatmap(role_attr_matrix, annot=True, fmt='d', cmap='YlOrRd',
            xticklabels=['str', 'agi', 'int', 'all'],
            yticklabels=top_5_roles['Role'], ax=ax19)
ax19.set_title('Распределение ролей по атрибутам', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

fig5 = plt.figure(figsize=(16, 10))

attr_colors = {'str': '#ff6b6b', 'agi': '#4ecdc4', 'int': '#45b7d1', 'all': '#96ceb4'}

for attr in attr_colors:
    subset = df_viz[df_viz['Primary Attribute'] == attr]
    plt.scatter(subset['Times Picked'], subset['Win Rate'],
                s=subset['Total Pro wins']*2, alpha=0.6,
                c=attr_colors[attr], label=attr, edgecolors='black', linewidth=0.5)

top_heroes = df_viz.nlargest(15, 'Total Pro wins')
for _, hero in top_heroes.iterrows():
    plt.annotate(hero['Name'], (hero['Times Picked'], hero['Win Rate']),
                fontsize=9, fontweight='bold', ha='center', va='bottom')

plt.xlabel('Количество пиков', fontsize=12)
plt.ylabel('Win Rate (%)', fontsize=12)
plt.title('Пузырьковая диаграмма: Пики vs Win Rate (размер = победы)',
          fontsize=16, fontweight='bold')
plt.legend(title='Атрибут')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("Статы")
print("="*60)
print(f"Всего героев: {len(df_viz)}")
print(f"Средний Win Rate: {df_viz['Win Rate'].mean():.2f}%")
print(f"Максимальный Win Rate: {df_viz['Win Rate'].max():.2f}% (герой: {df_viz.loc[df_viz['Win Rate'].idxmax(), 'Name']})")
print(f"Минимальный Win Rate: {df_viz['Win Rate'].min():.2f}% (герой: {df_viz.loc[df_viz['Win Rate'].idxmin(), 'Name']})")
print(f"Всего побед на про-сцене: {df_viz['Total Pro wins'].sum():,}")
print(f"Всего пиков: {df_viz['Times Picked'].sum():,}")
print(f"Всего банов: {df_viz['Times Banned'].sum():,}")
```
